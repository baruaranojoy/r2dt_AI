```python
from bs4 import BeautifulSoup
import requests
import time
import os
import urllib.request, json 
with urllib.request.urlopen("https://newsapi.org/v2/top-headlines?sources=bbc-news&apiKey=4d18fcdb680b4d8baac9709e2b2ec307") as url:
    dirc=os.getcwd()
    data = json.loads(url.read().decode())
    #print(data)
    ts = int(time.time())
    total=data["totalResults"]
    for i in range(0,total):
        x=data["articles"][i]["url"]
        page = requests.get(x)
        soup = BeautifulSoup(page.content, 'html.parser')
    #st=soup.find(id="page")
        try:
            stry=soup.find_all(class_='story-body')
            stry_bdy=soup.find(class_='story-body__inner')
            x=soup.find_all('p', class_='story-body__introduction')
            abc = stry_bdy.find_all('p')
            periods = [a.get_text() for a in abc]
            filename=str(dirc)+'/Data/bbc_'+str(i)+'_'+str(ts)+'.txt'
            thefile = open(filename, 'w')
            for item in periods:
              thefile.write(str(item)+"")
              print(item)
            thefile.close()
        except: pass  
```
