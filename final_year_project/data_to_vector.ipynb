```python
from gensim.test.utils import common_texts
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import multiprocessing
from gensim.test.utils import get_tmpfile
```

```python
fopen = open("corpus/corpus_pool.txt")
data = fopen.read()
data = data.split("\n")
data = data[:len(data) - 1]

for k in range(len(data)):
    data[k] = data[k].split(" ")

documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(data)]
```

```python
model = Doc2Vec(documents,
                vector_size = 200,
                window = 5,
                min_count = 1,
                workers = multiprocessing.cpu_count(),
                sample = 1e-5, seed = 1)
```

```python
fname = "corpus/custom_doc2vec_model"
model.save(fname)
```

```python
model__ = Doc2Vec.load(fname)  # you can continue training with the loaded model!
vector = model__.infer_vector(data[5784])
print (vector)
```
