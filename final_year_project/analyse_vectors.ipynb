```python
import numpy as np
import glob
import os
import csv

from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
source_real = "Vectors/Real_consize/"
source_fake = "Vectors/Fake_consize/"
```

```python
def getall_vec(path):
    toret = []
    allfile_name = glob.glob(path + "*.csv")
    for k in range(len(allfile_name)):
        fopen = open(allfile_name[k], "r")
        data = fopen.read()
        data = data.split("\n")
        data = data[:len(data) - 1]
        zero = 10 - len(data)
        
        if len(data) + zero != 10:
            print (len(data) + zero)
        
        for m in range(len(data)):
            data[m] = data[m].split(",")
        
        for p in range(len(data)):
            for q in range(len(data[p])):
                data[p][q] = float(str(data[p][q]))    
        
        for a in range(len(data)):
            toret.append(data[a])
        
        for d in range(zero):
            ol = [0]*200
            toret.append(ol)
            
    return toret
```

```python
def getall_vecfile(path):
    toret = []
    allfile_name = glob.glob(path + "*.csv")
    for k in range(len(allfile_name)):
        fopen = open(allfile_name[k], "r")
        data = fopen.read()
        data = data.split("\n")
        data = data[:len(data) - 1]
        zero = 10 - len(data)
        
        if len(data) + zero != 10:
            print (len(data) + zero)
        
        for m in range(len(data)):
            data[m] = data[m].split(",")
        
        for p in range(len(data)):
            for q in range(len(data[p])):
                data[p][q] = float(str(data[p][q]))    
        
        for a in range(len(data)):
            toret.append(data[a])
        
        for d in range(zero):
            ol = [0]*200
            toret.append(ol)
            
    to_ret = []
    for k in range(int(len(toret)/10)):
        dop = []
        for m in range(10):
            for l in range(len(toret[k*10 + m])):
                dop.append(toret[k*10 + m][l])
        to_ret.append(dop)
    return to_ret
```

# Load real and fake data and plot in 2D (sentence wise)

```python
real_vectors = getall_vec(source_real)
fake_vectors = getall_vec(source_fake)
```

```python
print (len(real_vectors), len(fake_vectors))
```

```python
tsne = TSNE(n_components=2, random_state=0)

x_test_2d = tsne.fit_transform(real_vectors)
for k in range(len(x_test_2d)):
    plt.plot(x_test_2d[k][0], x_test_2d[k][1], 'r*')

x_test_2d = tsne.fit_transform(fake_vectors)
for k in range(len(x_test_2d)):
    plt.plot(x_test_2d[k][0], x_test_2d[k][1], 'b*')

plt.show()
```

# Interactive 3D plot (sentence wise)

```python
import numpy as np
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot,iplot

import plotly.plotly as py
import plotly.graph_objs as go
```

```python
real_vectors = getall_vec(source_real)
fake_vectors = getall_vec(source_fake)
```

```python
tsne = TSNE(n_components=3, random_state=0)
real_3d = tsne.fit_transform(real_vectors)
fake_3d = tsne.fit_transform(fake_vectors)
```

```python
real_x = []
real_y = []
real_z = []

for k in range(len(real_3d)):
    real_x.append(real_3d[k][0])
    real_y.append(real_3d[k][1])
    real_z.append(real_3d[k][2])
    

fake_x = []
fake_y = []
fake_z = []

for k in range(len(fake_3d)):
    fake_x.append(fake_3d[k][0])
    fake_y.append(fake_3d[k][1])
    fake_z.append(fake_3d[k][2])
```

```python
trace1 = go.Scatter3d(x = real_x, 
                      y = real_y, 
                      z = real_z,
                      mode = 'markers',
                      marker = dict( color='rgb(255, 0, 0)',
                                     size=12,
                                     line=dict(
                                     color='rgba(255, 255, 255, 0.0)',
                                     width=0.5 ),
                      opacity=0.8 ))
```

```python
trace2 = go.Scatter3d(x = fake_x, 
                      y = fake_y, 
                      z = fake_z,
                      mode = 'markers',
                      marker = dict( color='rgb(0, 0, 255)',
                                     size=12,
                                     line=dict(
                                     color='rgba(255, 255, 255, 0.0)',
                                     width=0.5 ),
                      opacity=0.8 ))
```

```python
data = [trace1, trace2]
layout = go.Layout( margin=dict(l = 0,
                                r = 0,
                                b = 0,
                                t = 0 ))

fig = go.Figure(data=data, layout=layout)
plot(fig,show_link = False)
```
